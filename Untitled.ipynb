```python
import os
import pydicom
import torch
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
folders = ['./input','./input2','./input3','./input4']
```

```python
images = []
inputs = []
optim = []
```

```python
def preprocess_file(file):
    with open(file,'r') as f:
        output = [[int(num) for num in line.split(' ')] for line in f]
    return output
```

```python
for x in folders:
    y = os.listdir(x)
    for file in y:
        if 'png' not in file:
            if 'HU' in file:
                images.append(preprocess_file(os.path.join(x,file)))
            elif 'seg' in file:
                inputs.append(preprocess_file(os.path.join(x,file)))
            elif 'opt' in file:
                optim.append(preprocess_file(os.path.join(x,file)))
```

```python
images = np.array(images)
inputs = np.array(inputs)
optim = np.array(optim)
```

```python
def display_image(example,grid):
    matrix = []
    for i in range(example.shape[0]):
        temp = []
        for j in range(example.shape[1]):
            if grid[i][j] == False:
                temp.append(0)
            else:
                temp.append(example[i][j])
        matrix.append(temp)
    matrix = np.array(matrix)
    return matrix
```

```python
def get_contour(drawing):
    matrix =[]
    for x in range(drawing.shape[0]):
        for y in range(drawing.shape[1]):
            if drawing[x][y] == 1:
                matrix.append([x,y])
    return np.array(matrix)
def check_image(example):
    for x in range(200,400,40):
        #red = [0,0,255]
        proceesed_image = display_image(example,(40<example) & (example<x))
        image_2d_scaled = (np.maximum(proceesed_image,0) / proceesed_image.max()) * 255.0
        image_2d_scaled = np.uint8(image_2d_scaled)
        color = cv.cvtColor(image_2d_scaled, cv.COLOR_GRAY2BGR) 
        
        cv.imshow('CT SCAN', color) 
        cv.waitKey(0) 
        cv.destroyAllWindows()
def getting_pixel_values(contur,imagine):
    valori = set()
    for puncte in contur:
        if imagine[puncte[0],puncte[1]] >=52:
            valori.add((imagine[puncte[0],puncte[1]]))
    return valori
def draw_contour(contur,imagine):
    proceesed_image = display_image(imagine,(40<imagine) & (imagine<200))
    image_2d_scaled = (np.maximum(proceesed_image,0) / proceesed_image.max()) * 255.0
    image_2d_scaled = np.uint8(image_2d_scaled)
    color = cv.cvtColor(image_2d_scaled, cv.COLOR_GRAY2BGR) 
    red = [0,0,255]
    for x in contur:
        if image_2d_scaled[x[0],x[1]] >=52:
            color[x[0],x[1]] = red
    cv.imshow('CT SCAN', color) 
    cv.waitKey(0) 
    cv.destroyAllWindows()
```
